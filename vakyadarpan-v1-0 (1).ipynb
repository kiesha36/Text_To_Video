##### <span style="font-family:cursive;"><h1><b>[4.3] Delimiter based text segmentation and pairing using List of Dictionaries</b></h1></span>

```python
sentence_info_list = []
if language_code=='en': #For English
    from nltk.tokenize import sent_tokenize
    sentences = sent_tokenize(summary)
    for idx, sentence in enumerate(sentences, start=1):
        sentence_info = {
            'id': idx,      
            'sentence': sentence,
            'image': None,  
            'audio': None,
            'video': None
        }
        sentence_info_list.append(sentence_info)
    for sentence_info in sentence_info_list:
        print(sentence_info)
elif language_code=='ur': #For Urdu
    sentence_split_pattern = r'[؟۔!…]'
    sentences = re.split(sentence_split_pattern, translated_summary)
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
    for idx, sentence in enumerate(sentences, start=1):
        sentence_info = {
            'id': idx,
            'sentence': sentence,
            'image': None,
            'audio': None,
            'video': None
        }
        sentence_info_list.append(sentence_info)
    for sentence_info in sentence_info_list:
        print(sentence_info)
else:#For rest of the Indic languages
    sentences = translated_summary.split('।')
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
    for idx, sentence in enumerate(sentences, start=1):
        sentence_info = {
        'id': idx,
        'sentence': sentence,
        'image': None,
        'audio': None,
        'video': None
        }
        sentence_info_list.append(sentence_info)
    for sentence_info in sentence_info_list:
        print(sentence_info)
```

##### <span style="font-family:cursive;"><h1><b>[6.2] Assigning an image from Database to each idx in List of Dictionaries</b></h1></span>

```python
!pip install random
```

```python
import random
def assign_images(sentence_info_list, db_images):
    db_images_copy = list(db_images)
    random.shuffle(db_images_copy)
    assigned_images = []
    for sentence_info in sentence_info_list:
        if len(db_images_copy) == 0:
            continue
        image = db_images_copy[0]
        if len(assigned_images) > 0 and assigned_images[-1] == image:
            db_images_copy.remove(image)
            continue
        assigned_images.append(image)
        sentence_info['image'] = path+image
    return sentence_info_list
sentence_info_list = assign_images(sentence_info_list, db_images)
```

##### <span style="font-family:cursive;"><h1><b>[7.2] Importing the dependencies and loading the files</b></h1></span>

```python
import moviepy.editor as mpe
import math
import numpy as np
```

##### <span style="font-family:cursive;"><h1><b>[7.3] Interpolation process</b></h1></span>

```python
def generate_video(sentence_info):
    image = mpe.ImageClip(sentence_info['image'])
    image = image.resize((1920, 1080))
    audio_clip = mpe.AudioFileClip(sentence_info['audio'])
    image.duration = audio_clip.duration
    initial_zoom = 1.0
    final_zoom = 1.5
    zoom_scales = np.linspace(initial_zoom, final_zoom, int(float(image.duration) * 30))
    zooming_frames = []
    for scale in zoom_scales:
        frame = image.resize(scale)
        frame = frame.set_position(("center", "center"))
        frame = frame.set_duration(1 / 30)
        zooming_frames.append(frame)
    zooming_clip = mpe.concatenate_videoclips(zooming_frames, method="compose")
    zooming_clip.audio = audio_clip
    return zooming_clip
```

##### <span style="font-family:cursive;"><h1><b>[7.4] Generating video for each idx in List of Dictionaries using Interpolation process </b></h1></span>

```python
for sentence_info in sentence_info_list:
    video_clip = generate_video(sentence_info)
    sentence_info['video'] = video_clip
```

##### <span style="font-family:cursive;"><h1><b>[7.5] Merge Process</b></h1></span>

```python
video_clips = []
for sentence_info in sentence_info_list:
    video_clip = sentence_info['video']
    video_clips.append(video_clip)
concatenated_clip = mpe.concatenate_videoclips(video_clips)
concatenated_clip.write_videofile("output_final_video.mp4")
```
